```python
import time
import os
import numpy as np
from ultralytics import YOLO

model = YOLO("runs/detect/train-3/weights/best.pt")

images = os.listdir("dataset/test/images")

times = []

for img in images:

    start = time.time()

    model.predict(
        source=f"dataset/test/images/{img}",
        verbose=False
    )

    finish = time.time()

    times.append(finish - start)

print(f"Среднее время: {np.mean(times):.4f} сек")
```

```python
import os
import time
import numpy as np
from ultralytics import YOLO

model = YOLO("runs/detect/train-3/weights/best.pt")

images = os.listdir("dataset/test/images")

# Прогрев модели
model.predict(
    source=f"dataset/test/images/{images[0]}",
    verbose=False
)

times = []

for img in images:

    start = time.perf_counter()

    model.predict(
        source=f"dataset/test/images/{img}",
        verbose=False
    )

    finish = time.perf_counter()

    times.append(finish - start)

print(f"Среднее время обработки: {np.mean(times):.4f} сек")
print(f"FPS: {1/np.mean(times):.2f}")
```
