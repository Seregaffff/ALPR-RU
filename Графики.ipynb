```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("runs/detect/train-3/results.csv")

plt.plot(df["epoch"], df["metrics/mAP50(B)"])
plt.title("mAP50 YOLO training")
plt.xlabel("epoch")
plt.ylabel("mAP50")
plt.show()

plt.plot(df["epoch"], df["metrics/precision(B)"])
plt.title("Precision")
plt.show()

plt.plot(df["epoch"], df["metrics/recall(B)"])
plt.title("Recall")
plt.show()
```
